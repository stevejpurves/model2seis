```python
import sys, os
import matplotlib.pyplot as plt
import numpy as np
# adjust some settings for matplotlib
from matplotlib import rcParams
# print rcParams
rcParams['font.size'] = 15
import pynoddy
import pynoddy.history
import pynoddy.output
```

```python
def computeAndDisplaySection(history_file, output_name, cube_size):
    # We will first recompute the model and store results in an output file for comparison
    NH1 = pynoddy.history.NoddyHistory(history_file)
    NH1.change_cube_size(cube_size)
    NH1.write_history(history_file+'.mod')
    pynoddy.compute_model(history_file+'.mod', output_name)
    NO1 = pynoddy.output.NoddyOutput(output_name)
    # create basic figure layout
    fig = plt.figure(figsize = (15,5))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    NO1.plot_section('x', position=10, ax=ax1, colorbar=True, title=output_name+',x')
    NO1.plot_section('y', position=10, ax=ax2, colorbar=True, title=output_name+',y')
    section, pos = NO1.get_section_voxels('y', position=10)
    print(np.unique(section))
    plt.show()
```

## Examples with Stochastic Parameter files
That we can run multiple reaslisations on

```python
computeAndDisplaySection('fold_dyke_fault.his', 'fold_dyke_fault', 25)
```

```python
computeAndDisplaySection('GBasin_simplified.his','GBasin_simplified',50)
```
