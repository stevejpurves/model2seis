```python
models = ["fold_dyke_fault","GBasin_simplified","slightly_complicated","simple_two_faults","two_faults_fold_unconformity_slice","GBasi123","GBasin_Ve1_V4","GBasin_Ve1_V4_b","GBasin_Ve1_V4_ve10","GBasin_Ve1_V5","test1","test2","test_basin","test_thrust"]

def event_classifier(models,verbose=True):
    eventid={}
    for modelname in models:
        if verbose: print("\nModel: "+modelname)
        H1 = pynoddy.history.NoddyHistory(modelname+".his")
        faultsaver=[]
        dykesaver=[]
        tiltsaver=[]
        ucsaver=[]
        shearsaver=[]
        foldsaver=[]
        for noddies in range(1,len(H1.events)+1):
            if "FAULT" in H1.events[noddies].event_type:
                faultsaver.append(noddies)
            if "DYKE" in H1.events[noddies].event_type:
                dykesaver.append(noddies)
            if "TILT" in H1.events[noddies].event_type:
                tiltsaver.append(noddies)
            if "UNCONFORMITY" in H1.events[noddies].event_type:
                ucsaver.append(noddies)
            if "SHEAR" in H1.events[noddies].event_type:
                shearsaver.append(noddies)
            if "FOLD" in H1.events[noddies].event_type:
                foldsaver.append(noddies)
            #print(H1.events[noddies].event_type)
        eventid[modelname] = {}
        eventid[modelname]['fault'] = faultsaver
        eventid[modelname]['dyke'] = dykesaver
        eventid[modelname]['unconformity'] = ucsaver
        eventid[modelname]['shear'] = shearsaver
        eventid[modelname]['tilt'] = tiltsaver
        eventid[modelname]['fold'] = foldsaver
        
        if verbose:
            print(str(len(faultsaver))+" Faults identified")
            print(str(len(dykesaver))+" Dykes identified")
            print(str(len(tiltsaver))+" Tilts identified")
            print(str(len(ucsaver))+" Unconformities identified")
            print(str(len(shearsaver))+" Shearzones identified")
            print(str(len(foldsaver))+" Folds identified")
    return eventid


```

Unconformity: x, y, z, dip, dip direction (5)

```python
def pandafy(modelobject,event):
    out = pd.DataFrame(columns=["event","parameter","type","mean","+-","name"])
    for key, value in event.iteritems():
        if key == "fault":
            parameters = ["Dip","Dip Direction","Slip"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Slip":
                        paramodifier=3
                    else:
                        paramodifier=1
                    if x == "Dip" and mean_val == 0:
                        std_val = 5
                    else:
                        std_val = min(abs(int(np.ceil(.15*mean_val*paramodifier))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"vonmises","mean":mean_val,"+-":std_val,"name":key}
        if key == "tilt":
            parameters = ["Plunge", "Plunge Direction", "Rotation"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Rotation":
                        paramodifier=.5
                    else:
                        paramodifier=1
                    std_val = min(abs(int(np.ceil(.15*mean_val*paramodifier))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"normal","mean":mean_val,"+-":std_val,"name":key}
        if key == "fold":
            parameters = ["Wavelength", "Amplitude"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Amplitude":
                        paramodifier=2
                    else:
                        paramodifier=1
                    std_val = min(abs(int(np.ceil(.15*mean_val*paramodifier))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"normal","mean":mean_val,"+-":std_val,"name":key}
        if key == "unconformity":
            parameters = ["Dip","Dip Direction"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Dip" and mean_val == 0:
                        std_val = 5
                    else:
                        std_val = min(abs(int(np.ceil(.15*mean_val))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"normal","mean":mean_val,"+-":std_val,"name":key}

        if key == "shear":
            parameters = ["Dip","Dip Direction","Slip"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Slip":
                        paramodifier=3
                    else:
                        paramodifier=1
                    if x == "Dip" and mean_val == 0:
                        std_val = 5
                    else:
                        std_val = min(abs(int(np.ceil(.15*mean_val*paramodifier))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"normal","mean":mean_val,"+-":std_val,"name":"shear zone"}
        if key == "dyke":
            parameters = ["Dip","Dip Direction"]
            for y in value:
                for x in parameters:
                    mean_val = modelobject.events[y].properties[x]
                    if x == "Dip" and mean_val == 0:
                        std_val = 5
                    else:
                        std_val = min(abs(int(np.ceil(.15*mean_val))),abs(int(.9*mean_val)))
                    out.loc[len(out.index)+1]={"event":y,"parameter":x,"type":"vonmises","mean":mean_val,"+-":std_val,"name":key}

    return out
        
```
